# 1.1 Image Color with different libraries

#  1.3 Reduce Bitdepth using Quantization

```python

fig2 = plt.figure(figsize=(10, 7))

default_img2 = cv2.imread('18in.JPG')
gray = cv2.cvtColor(default_img2, cv2.COLOR_BGR2GRAY)

fig2.add_subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.show()

bit_depth = 3

qlevel = 2**bit_depth  #2^3

# min=0  , max = 255 
smin = 0
smax = 255
fig2.add_subplot(1,2,2)
gray = ((gray-smin)/(smax-smin))*qlevel
gray = gray.astype(np.uint8)
plt.imshow(gray, cmap='gray')
plt.show()

```

# 1.4  3D Image Surface

```python
gray_img = cv2.imread('76937540_1281214988716549_578073933429866496_n.jpg',0)
imgresize = cv2.resize(gray_img, (300,300))

gridx,gridy = np.mgrid[0:300,0:300]
fig = plt.figure()

plot3d = fig.add_subplot(111, projection='3d')
plot3d.plot_surface(gridx, gridy, imgresize, cmap='gray')
plt.show()
```
